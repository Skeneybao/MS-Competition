```python
import pandas as pd
```

```python
key="GoldmanSachs"
numbers='10000'
start_dates=[str(year)+'-'+str(month)+'-01' for year in ["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"] for month in ["{0:0=2d}".format(m) for m in range(1,13)]]
end_dates=start_dates.copy()
for i,d in enumerate(end_dates):
    if d[5:7] in ['01','03','05','07','08','10','12']:
        end_dates[i]=d[:-2]+'31'
    elif d[5:7] in ['04','06','09','11']:
        end_dates[i]=d[:-2]+'30'
    else:
        end_dates[i]=d[:-2]+'28'
        
print(end_dates)
```

```python
for i in range(len(start_dates)):
    name1=key+start_dates[i][:7]+'.json'
    name2='pretty_'+name1
    print('twitterscraper '+key+' -bd '+start_dates[i]+' -ed '+end_dates[i]+' -o '+name1+' -l '+numbers)
    print("sleep 900s")
    #print('python -m json.tool '+name1+' > '+name2)
```
