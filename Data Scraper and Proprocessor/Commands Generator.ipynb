```python
import pandas as pd
```

```python
key="TDAmeritrade"
numbers='10000'
start_dates=[str(year)+'-'+str(month)+'-01' for year in ["2017","2018","2019","2020"] for month in ["{0:0=2d}".format(m) for m in range(1,13)]]
             #["2010","2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"] for month in ["{0:0=2d}".format(m) for m in range(1,13)]]
end_dates=start_dates.copy()
end_dates.pop(0)
end_dates.append('end')
        
print(start_dates)
```

```python
daylist=['01','08','15','22']
for i in range(len(start_dates)):
    for j in range(4):
        name1=key+start_dates[i][:7]+'-{}.json'.format(j+1)
        bd=start_dates[i][:-2]+daylist[j]
        if j!=3:
            ed=start_dates[i][:-2]+daylist[j+1]
        else:
            ed=end_dates[i]
        print('twitterscraper '+key+' -bd '+ bd +' -ed '+ed+' -o '+name1+' -l '+numbers)
        print("sleep 900s")
    #print('python -m json.tool '+name1+' > '+name2)
```
