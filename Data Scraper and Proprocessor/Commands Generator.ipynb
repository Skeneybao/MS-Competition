```python
key="private_wealth"
numbers='10000'
start_dates=[str(year)+'-'+str(month)+'-01' for year in ["2015","2016","2017","2018","2019"] for month in ["{0:0=2d}".format(m) for m in range(1,13)]]
end_dates=start_dates.copy()
end_dates.append("2020-01-01")
end_dates.pop(0)
print(end_dates)
```
