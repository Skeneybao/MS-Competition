```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
params = {'legend.fontsize': 'xx-large',
          'figure.figsize': (15, 5),
         'axes.labelsize': 'xx-large',
         'axes.titlesize':'xx-large',
         'xtick.labelsize':'xx-large',
         'ytick.labelsize':'xx-large'}
pylab.rcParams.update(params)
```

```python
data = np.array([[16,73, 34, 46, 29],
[84,27, 66, 54, 71]])
data2=np.array([45,720,110,50,30])
X = np.arange(5)
fig = plt.figure(figsize=(9,6))
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0]/100, color = 'plum', width = 0.25)
ax.bar(X + 0.25, data[1]/100, color = 'aquamarine', width = 0.25)
ax.plot(X,data2/sum(data2), linewidth=10,color='cornflowerblue')
#ax.plot(X,(data[0]-data[1]+100)/200)
ax.set_xticks(np.arange(5))
ax.set_xticklabels(['2020-01','2020-02','2020-03','2020-04','2020-05'])
fig.savefig('test.jpg')
```

```python
def r():
    return np.abs(np.random.randn())
data = np.array([[8+r(), 3+r(), 1+r(), 5+r(),1+r(),3+r()],                 
                [1+r(), 1+r(), 4+r(), 3+r(),1+r(),1+r()],
                [1+r(),4+r(),3+r(),r(),r(),r()]]
               )
X = np.arange(6)
fig = plt.figure(figsize=(9,6))
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0]/100, color = 'plum', width = 0.25,label='Phone Service')
ax.bar(X + 0.25, data[1]/100, color = 'aquamarine', width = 0.25,label='Transfer Issues')
ax.bar(X + 0.5, data[2]/100, color = 'cornflowerblue', width = 0.25,label='No chat/No Respond')
#ax.plot(X,data2/sum(data2), linewidth=10,color='cornflowerblue')
#ax.plot(X,(data[0]-data[1]+100)/200)
ax.set_xticks(np.arange(6))
ax.set_xticklabels(['etrade','TDA','Robin','Wells','CS','Merrill'])
ax.legend(loc='best')
fig.savefig('Customer Services')
```
